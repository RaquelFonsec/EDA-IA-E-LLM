```python


data = {
    'matricula': [1, 2, 3, 4, 5],
    'nome': ['Lucas', 'Ana', 'Luiza', 'Fernando', 'Sandra'],
    'cidade': ['Atibaia', 'São Paulo', 'Santos', 'Atibaia', 'Atibaia'],
    'estado': ['SP', 'SP', 'SP', 'SP', 'SP'],
    'pais': ['Brasil', 'Brasil', 'Brasil', 'Brasil', 'Brasil'],
    'idade': [35, 29, 38, 36, 28],
    'departamento': ['Compras', 'Vendas', 'Finanças', 'Marketing', 'Produção'],
    'cargo': ['Gerente', 'Coordenador', 'Gerente', 'Diretor', 'Analista'],
    'salario': [25000, 12000, 28000, 40000, 23000],
    'escolaridade': ['Superior', 'Superior', 'MBA', 'Mestrado', 'Superior'],
    'nota': [8, 6, 9, 7, 5],
    'sexo': ['M', 'F', 'F', 'M', 'F']
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Configurações para gráficos
sns.set(style="whitegrid")

# Gráfico de contagem de sexo
plt.figure(figsize=(8, 6))
sns.countplot(x='sexo', data=df, hue='sexo', palette='Set2', legend=False)
plt.title('Contagem de Funcionários por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.xticks([0, 1], ['Masculino', 'Feminino'])
plt.show()

# Gráfico de contagem de cidades
plt.figure(figsize=(8, 6))
sns.countplot(x='cidade', data=df, hue='cidade', palette='Set1', legend=False)
plt.title('Contagem de Funcionários por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()

# Gráfico de contagem de escolaridade
plt.figure(figsize=(8, 6))
sns.countplot(x='escolaridade', data=df, hue='escolaridade', palette='Set3', legend=False)
plt.title('Contagem de Funcionários por Escolaridade')
plt.xlabel('Escolaridade')
plt.ylabel('Contagem')
plt.show()

```

```python
# Distribuição de Idades
plt.figure(figsize=(8,5))
sns.histplot(df['idade'], bins=20, kde=True, color='blue')
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Contagem (count): O conjunto de dados contém 10 registros de idade.
Média (mean): A idade média dos funcionários é de 32,4 anos.
Desvio Padrão (std): A variação das idades é de aproximadamente 4,4 anos, o que indica uma distribuição moderadamente dispersa.
Mínimo (min): A menor idade registrada é 28 anos.
1º Quartil (25%): 25% dos funcionários têm idade inferior a 29 anos.
Mediana (50%): A idade mediana é 30 anos, ou seja, metade dos funcionários têm idade abaixo de 30 anos e metade acima.
3º Quartil (75%): 75% dos funcionários têm idade inferior a 35,75 anos.
Máximo (max): A maior idade registrada é 40 anos.



Resumo:


A faixa etária dos funcionários varia de 28 a 40 anos, com a maioria (50%) dos funcionários tendo idades entre 29 e 35,75 anos. A média de idade é de 32,4 anos, o que reflete um grupo de funcionários relativamente jovem, mas com uma variação considerável nas idades.

```python
# Analisando Salários por Cargo
plt.figure(figsize=(10,6))
sns.boxplot(x='cargo', y='salario', data=df, palette='viridis', hue='cargo', legend=False)
plt.xticks(rotation=45)
plt.title('Salário por Cargo')
plt.show()

```

Em geral, Diretores têm o maior salário fixo, enquanto Analistas apresentam maior variação salarial.

```python
# Selecionando apenas colunas numéricas para correlação
df_numeric = df.select_dtypes(include=['number'])

# Correlação entre variáveis numéricas
plt.figure(figsize=(8,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação')
plt.show()

print("Análise Exploratória concluída!")
```

Matrícula e Idade: A correlação é de -0.04, indicando uma relação muito fraca e quase nenhuma relação linear entre essas variáveis.



Matrícula e Salário: A correlação é de -0.54, sugerindo uma correlação negativa moderada entre a matrícula e o salário, ou seja, conforme o número da matrícula aumenta, o salário tende a diminuir, embora a relação não seja muito forte.


Matrícula e Nota: A correlação é de -0.55, indicando uma correlação negativa moderada entre a matrícula e a nota, ou seja, conforme o número da matrícula aumenta, a nota tende a diminuir.




Idade e Salário: A correlação é de 0.46, indicando uma correlação moderada positiva entre a idade e o salário. Isso sugere que, à medida que a idade aumenta, o salário tende a aumentar, mas a relação não é muito forte.


Idade e Nota: A correlação é de 0.52, sugerindo uma correlação moderada positiva entre a idade e a nota, indicando que funcionários mais velhos tendem a ter notas um pouco mais altas.



Salário e Nota: A correlação é de 0.42, indicando uma correlação positiva moderada entre o salário e a nota. Isso sugere que salários mais altos estão levemente associados a notas mais altas.



Resumo:


A matriz de correlação indica que as variáveis numéricas têm algumas correlações moderadas, com destaque para a relação entre idade e salário (positiva) e entre salário e nota (também positiva). No entanto, as correlações mais fracas estão entre matrícula e as outras variáveis.
